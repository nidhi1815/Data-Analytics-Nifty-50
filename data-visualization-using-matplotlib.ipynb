```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\nidhi\OneDrive\Desktop\projects\DA\Toyota.csv.csv"
cars_data = pd.read_csv(file_path)

plt.hist(cars_data['KM'], color='green', edgecolor='white', bins=10)
plt.title("Histogram of Kilometer")
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

counts = [979, 120, 12,40]  # counts of each category
fuelType = ('Petrol', 'Diesel', 'CNG','EV')
index = np.arange(len(fuelType))  # index = fuelType , count
plt.bar(index, counts, color=['red', 'blue', 'cyan','green'])

plt.xlabel('Fuel Type')
plt.ylabel('Frequency')
plt.title('Bar Graph of Fuel Type')
plt.xticks(index, fuelType, rotation=60)

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\nidhi\OneDrive\Desktop\projects\DA\Toyota.csv.csv"
df = pd.read_csv(file_path)


fuel_counts = df['FuelType'].value_counts()
fuel_counts.plot.bar(color=['blue', 'green', 'orange'] , edgecolor= 'black')


plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.title('Bar Graph of Fuel Type')

plt.show()
```

```python
#line chart

import pandas as pd
import matplotlib.pyplot as plt

year =[1920,1930,1940,1950,1960,1970,1980,1990,2000,2010]
unemployement_rate=[9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,6.3]

plt.plot(year ,unemployement_rate, color='red',marker='o' )
plt.xlabel('Year')
plt.ylabel('Unemployement rate')
plt.title('Unemployement rate v/s year line graph', fontsize=20)

plt.grid(True) #grid laane k liye graph me

plt.show()
```
